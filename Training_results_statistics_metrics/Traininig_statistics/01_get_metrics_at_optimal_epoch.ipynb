```python

def get_metrics_dataframe(
          optimal_epoch_df_path,
          results_path, 
          training_type,
          metric_name,
          output_dir
):
    
    if training_type == "finetuning":
        training_logs_file_name = "trainer_state.json"
    if training_type == "pet":
        training_logs_file_name = "results.json"

    metrics_grid = {}

    for model_name, model_full_name in MODEL_NAME_TO_FULL_NAME.items():


        

        for content_source, content_source_full_name in CONTENT_NAME_TO_FULL_NAME.items():

            # get content source directory
            content_source_dir = get_string_with_prefixes(
                content_source_full_name,
                os.path.join(
                results_path, 
                #"finetuning_results"
                )
            )


            # get model subdirectory
            model_dir = get_string_with_prefixes(
                model_full_name,
                os.path.join(results_path, 
                             #"finetuning_results", 
                             content_source_dir)
            )
            #print(results_path, content_source_dir, model_dir, sep="\n")
            training_path = os.path.join(
                results_path, 
                #"finetuning_results", 
                content_source_dir,
                model_dir
            )

            if "auc" in metric_name.lower():
                if "dev" in metric_name:
                    split = "dev"
                if "test" in metric_name:
                    split = "test"
                training_logs_file_path = os.path.join(training_path, split, "performance_on_split", "metrics_on_split.json")

                with open(training_logs_file_path, "r") as f:
                    metrics_on_split = json.load(f)

                metric_value= metrics_on_split["AUC"]

            # get acc, precision, recall, f1, f2, f0.5 on the test split
            # expecting the metric_name to have the format
            # metric_name = "test_f_beta_2.0"
            if "test" in metric_name.lower():
                split = "test"
                training_logs_file_path = os.path.join(training_path, split, "performance_on_split", "metrics_on_split.json")

                with open(training_logs_file_path, "r") as f:
                    metrics_on_split = json.load(f)

                key = metric_name.replace("test", "").strip("_")
                metric_value= metrics_on_split[key]


            else:

                # load JSON file with results
                training_logs_file_path = os.path.join(training_path, "training_logs", training_logs_file_name)

                with open(training_logs_file_path, "r") as f:
                    
                    training_logs = json.load(f)
                    training_logs = training_logs["log_history"]

                # get optimal epoch
                optimal_epoch = get_value_from_df(optimal_epoch_df_path, content_source, model_name)
                
                # get metric at optimal epoch
                metric_value = get_metric_value_at_epoch(training_logs, optimal_epoch, metric_name)
                # print(metric_value, optimal_epoch, content_source, model_name)

            metrics_grid[
                (model_name, content_source)
            ] = metric_value


    ## construct output dataframe
    metrics_grid



    metrics_df = {
        "model": [
            MODEL_NAME_LIST_FOR_THESIS[name] for name in MODEL_NAME_LIST
            ]
        }

    # columns for content
    for content_source in CONTENT_NAME_TO_FULL_NAME.keys():

        metrics_df[content_source] = []

        for model_name in MODEL_NAME_TO_FULL_NAME.keys():

            value = metrics_grid[(model_name, content_source)]
            metrics_df[content_source].append(value)


    metrics_df = pd.DataFrame(metrics_df)

    # save df
    output_path = os.path.join(
        output_dir,
        f"{metric_name}__{os.path.basename(optimal_epoch_df_path)}" # already has .csv extension
    )

    metrics_df.to_csv(output_path)

    return metrics_df

    
            
```

```python
# needs to be executed after the previous commands, which build the dataframes


def compare_training_methods_for_model(
          metrics_summary_path,
          model_name, 
          metric_name,
          output_dir
):
    
    comparison_df = {"Content Source" : CONTENT_NAME_LIST}

    for summary_csv in os.listdir(metrics_summary_path):
            
        # is it the file we want ?

        if metric_name in summary_csv:

            # determine training method
            for method in TRAINING_METHODS_LIST:
                if f"{method} " in summary_csv:
                    training_method = TRAINING_METHODS_NAMES_FOR_THESIS[method]
                    print(summary_csv, training_method)
            

            metrics_df = pd.read_csv(os.path.join(metrics_summary_path, summary_csv))
            metrics_df = metrics_df.set_index("model").transpose()

            data = list(metrics_df[model_name])
            data.pop(0)
            # print(len(data), data)

            comparison_df[training_method] = data


    # print(comparison_df)
    comparison_df = pd.DataFrame(comparison_df)
    # reorder columns
    training_methods = [TRAINING_METHODS_NAMES_FOR_THESIS[name] for name in TRAINING_METHODS_LIST]
    comparison_df = comparison_df[["Content Source"] + training_methods]

    # print(comparison_df)

    # save df
    output_path = os.path.join(
        output_dir,
        f"{metric_name}__{model_name}.csv" 
    )

    comparison_df.fillna(0.0, inplace=True) # quick fix for retrieval error
    comparison_df.to_csv(output_path)

    return comparison_df


        
```
